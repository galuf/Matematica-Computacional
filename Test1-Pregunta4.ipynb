## Parte 1

```python
# modulos para graficar
from mpl_toolkits.mplot3d import Axes3D 
import matplotlib.pyplot as plt 

import numpy as np # para realizar algebra lineal
import math # opraciones trigonometricas
```

Sea un cubo ABCD-EFGH de la siguiente forma :


Generamos la estructura del cubo (cada vertice con los vertices que estara conectado)

```python
cubo_estructura = {
    "A" : ["B","D","E"],
    "B" : ["A","F","C"],
    "C" : ["B","D","G"],
    "D" : ["A","C","H"],
    "E" : ["A","F","H"],
    "F" : ["E","B","G"],
    "G" : ["C","F","H"],
    "H" : ["D","E","G"],
}
```

Definimos los vertices en el plano carteciano

```python
cubo_vertices = {
        "A" : [0,0,0], #[x,y,z] representacion de las coordenadas
        "B" : [0,1,0],
        "C" : [1,1,0],
        "D" : [1,0,0],
        "E" : [0,0,1],
        "F" : [0,1,1],
        "G" : [1,1,1],
        "H" : [1,0,1],
}
```

Teniendo los vertices y la estructura podemos generar una matriz con los puntos a graficar como linea

### Ejemplo:

la linea generada por el vertice A y el vertice B sera representado por una matriz del tipo

            0 0 0     -> coordenada de A
            0 1 0     -> coordenada de B
            
es matriz representa la arista AB

```python
def genera_estructura_aristas(cubo_vertices,cubo_estructura):
    
    cubo_aristas = []
    
    for i in cubo_estructura:
        for j in cubo_estructura[i]:
            m = []
            m.append(cubo_vertices[i]) #generamos la matriz con los dos vertices que generaran la linea
            m.append(cubo_vertices[j])
            cubo_aristas.append(m)
            
    # convertimos a numpy array para poder usar metodos como transpuesta        
    cubo_aristas = np.array(cubo_aristas)
    return cubo_aristas
```

Definimos la funcion dibujar la cual recorrera toda la matriz de aristas dibujando asi el cubo

```python
def dibuja(aristas):
    
    fig = plt.figure()
    ax = Axes3D(fig)
    
    for i in range(len(aristas)):
        #cada matriz esta representada como  0 0 0
        #                                    0 1 0
        #a esta matriz le sacamos la transpuesta para tener los valore de la coordenada como
        # x [0,0]  y = [0,1]  z = [0,0]
        ax.plot(aristas[i].T[0], aristas[i].T[1], aristas[i].T[2], marker='o', color='b')
        ax.set_xlabel('X') 
        ax.set_ylabel('Y') 
        ax.set_zlabel('Z')
    #mostrar la gráfia
    plt.show()
```

Dibujemos

```python
cubo_aristas = genera_estructura_aristas(cubo_vertices,cubo_estructura)

dibuja(cubo_aristas)
```

## Parte 2

Primero rotamos el eje Z en un angulo de pi/4 contra el reloj 

```python
angulo = math.pi/4

Rz = np.array([[math.cos(angulo),-math.sin(angulo),0],
               [math.sin(angulo),math.cos(angulo),0],
               [0,0,1]]) # counterclockwise
```

Ahora rotamos se debe rotar en un angulo alfa en el eje Y

```python
angulo1 = math.asin(math.sqrt(3)/3)
Ry = np.array([[math.cos(angulo1),0,-math.sin(angulo1)],
               [0,1,0],
               [math.sin(angulo1),0,math.cos(angulo1)]])
```

Finalmente rotamos pi/6 en sentido contra el reloj en el eje X

```python
angulo2 = math.pi/6
Rx = np.array([[1,0,0],
               [0,math.cos(angulo2),math.sin(angulo2)],
               [0,-math.sin(angulo2),math.cos(angulo2)]]) # counterclockwise
```

La funcion transforma tomara un vertice como coordenada [x,y,z] y le aplicara la transformacion

```python
def transforma(vector):
    m = []
    for i in vector:
        m.append([i])
    m = np.array(m)
    
    punto = np.dot( np.dot(np.dot(Rz,Ry),Rx) ,m)
    return list(punto.T[0]) 
```

Iteramos la transformacion para cada vertice (A,B,C,...)

```python
for i in cubo_vertices:
    cubo_vertices[i] = transforma(cubo_vertices[i])
```

Ahora graficamos el cubo con las coordenadas de los vertices transformados

```python
cubo_aristas_t = genera_estructura_aristas(cubo_vertices,cubo_estructura)

dibuja(cubo_aristas_t)
```
